#HOW TO DEAL WITH NULL VALUES

# Drop rows with any null values
data.dropna(axis=0, inplace=True)

# Drop columns with any null values
data.dropna(axis=1, inplace=True)

# Fill null values with mean of the column
data['column_name'].fillna(data['column_name'].mean(), inplace=True)

# Fill null values with median of the column
data['column_name'].fillna(data['column_name'].median(), inplace=True)

# Fill null values with mode of the column
data['column_name'].fillna(data['column_name'].mode()[0], inplace=True)

# Fill null values with a constant value
data['column_name'].fillna('Unknown', inplace=True)

# Forward fill null values (last known value)
data.fillna(method='ffill', inplace=True)

# Backward fill null values (next known value)
data.fillna(method='bfill', inplace=True)

When inplace = True , the data is modified in place, which means it will return nothing and the dataframe is now updated.

#DATA NORMALIZATION:
Used to map nominal data into uniform scale i.e. (0 to 1).
Also knowns as MIN-MAX SCALING

```python
from sklearn import preprocessing
```

```python
min_max_scaler=preprocessing.MinMaxScaler() #Creating a minimum and maximum processor object
```

```python
x=df.iloc[0:10,0:5]
x
```

```python
#Create an object to transform the data to fit minmax processor
x_scaled=min_max_scaler.fit_transform(x)
```

```python
#Run normalizer on dataframe
df_normalized=pd.DataFrame(x_scaled)
```

```python
df_normalized
```
