# Scan all numeric variables for outliers. If there are outliers, use any of the suitable techniques to deal with them.

```python
sns.distplot(df['math score']) #distplot is used to check data distribution
```

```python
df.boxplot(['math score','reading score','writing score'])
```

```python
#create a figure with multiple subplots and visualize the distribution and outliers of the 'math score' variable
plt.figure(figsize=(12,10))
plt.subplot(2,2,1)
df.boxplot(['math score'])
plt.subplot(2,2,2)
sns.kdeplot(df['math score'])
plt.subplot(2,2,3)
sns.histplot(df['math score'])
```

IQR (Inter Quartile Range) Inter Quartile Range approach to finding theoutliers is the most commonly used and most trusted approach used in the research field.

```python
#IQR=Q3-Q1
#upper = Q3 +1.5*IQR
#lower = Q1 – 1.5*IQR
IQR = df['math score'].quantile(0.75) - df['math score'].quantile(0.25)
lower_math = df['math score'].quantile(0.25) - 1.5 * IQR
df = df[(df['math score'] >= lower_math)]
```

```python
#Checking again if outliers are removed
plt.figure(figsize=(12,10))
plt.subplot(2,2,1)
df.boxplot(['math score'])
plt.subplot(2,2,2)
sns.kdeplot(df['math score'])
plt.subplot(2,2,3)
sns.histplot(df['math score'])
```

```python
##create a figure with multiple subplots and visualize the distribution and outliers of the 'reading score' variable
plt.figure(figsize=(12,10))
plt.subplot(2,2,1)
df.boxplot('reading score')
plt.subplot(2,2,2)
sns.histplot(df['reading score'])
plt.subplot(2,2,3)
sns.kdeplot(df['reading score'])
```

```python
IQR=df['reading score'].quantile(0.75)-df['reading score'].quantile(0.25)
lower_reading=df['reading score'].quantile(0.25)-1.5*IQR
df = df[(df['reading score'] >= lower_reading)]
```

```python
#Checking again if outliers are removed
plt.figure(figsize=(12,10)) #size of plots
plt.subplot(2,2,1) #coordinates of plots(row,column,position)
df.boxplot('reading score') #display of plot
plt.subplot(2,2,2)
sns.histplot(df['reading score'])
plt.subplot(2,2,3)
sns.kdeplot(df['reading score'])
```

```python
plt.figure(figsize=(12,7))
plt.subplot(2,2,1)
df.boxplot('writing score')
plt.subplot(2,2,2)
sns.histplot(df['writing score'])
plt.subplot(2,2,3)
sns.kdeplot(df['writing score'])
```
