```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=sns.load_dataset('titanic')
df
```

```python
df.columns
```

#Dataset Visualization

```python
sns.countplot(data=df,x='sex')
```

```python
sns.countplot(data=df,x='sex',hue='class')
```

```python
sns.catplot(x='sex',hue='survived',col='class',data=df,kind='count')
```

```python
sns.catplot(x='class',y='age',data=df,kind='boxen')
```

```python
sns.barplot(x='sex',y='survived',data=df,hue='class')
```

```python
sns.violinplot(x='sex',y='age',data=df,hue='survived')
```

```python
df.corr()
```

```python
plt.figure(figsize=(15,15))
sns.heatmap(df.corr(),annot=True);
#Coorelation
```
